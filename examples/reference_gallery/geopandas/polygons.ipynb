```python
import hvplot.pandas  # noqa
```

```python
import geopandas as gpd

countries = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
countries.sample(5)
```

```python
countries.hvplot()
```

```python
countries.hvplot.polygons(c='pop_est', hover_cols='all')
```

You can even color by another series, such as population density:

```python
countries.hvplot.polygons(c=countries.pop_est/countries.area, clabel='pop density')
```
